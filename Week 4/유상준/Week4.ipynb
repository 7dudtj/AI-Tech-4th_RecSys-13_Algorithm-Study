# 문제 1
- 순회강연
- https://www.acmicpc.net/problem/2109

- 전에 비슷한 유형을 풀었던 것 같다.
- 일단 내 리스트에 넣고, 길이가 초과하면 가장 작은애 버리기

```python
import heapq
import sys

# input = sys.stdin.readline
n = int(input())
tmp = [list(map(int,input().split())) for _ in range(n)] # List : [p,d]
tmp = sorted(tmp, key = lambda x : (x[1],-x[0]))

to_do = []
for p,d in tmp:
    heapq.heappush(to_do,p)
    while len(to_do) > d:
        heapq.heappop(to_do)
        
print(sum(to_do))
```
