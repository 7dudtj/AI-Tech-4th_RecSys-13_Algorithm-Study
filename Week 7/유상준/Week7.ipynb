# 문제 1
- 특정 거리의 도시 찾기
- https://www.acmicpc.net/problem/18352

```python
import sys
import heapq
# input = sys.stdin.readline

n,m,k,x = map(int,input().split())
graph = [[] for _ in range(n+1)]
for _ in range(m):
    a,b = map(int,input().split())
    graph[a].append(b)

tmp = [sys.maxsize]*(n+1)
tmp[x] = 0
q = []
heapq.heappush(q,(0,x))
while q:
    cost, c_node = heapq.heappop(q)
    if cost > k:
        break
    for n_node in graph[c_node]:
        if cost+1 < tmp[n_node]:
            tmp[n_node] = cost+1
            heapq.heappush(q,(cost+1, n_node))

if k not in tmp:
    print(-1)
else:
    for i in range(1,len(tmp)):
        if tmp[i] == k:
            print(i)
```
