## 문제 1 (공통 문제)
- 컵라면
- https://www.acmicpc.net/problem/1781

```python
from collections import defaultdict
import heapq
import sys

input = sys.stdin.readline
n = int(input())
tmp = []
for _ in range(n):
    d, k = map(int,input().split())
    tmp.append([d,k])
    
tmp.sort()

to_do = []
for d,k in tmp:
    heapq.heappush(to_do,k)
    if len(to_do) > d: # ex) 기한이 1인 과제인데, 이미 내가 할 과제리스트에 두 개 이상 들어있다면
        heapq.heappop(to_do)

print(sum(to_do))
```

## 문제 2
- 단어 수학
- https://www.acmicpc.net/problem/1339

```python
import heapq

n = int(input())

# 알파벳 별 도합 차수(클 수록 나중에 같은 위치에서 높은 숫자를 배정 받아야 함)
apb_cnt = {}        
# 알파벳 별 가지고 있는 숫자
for x in range(65,91):
    apb_cnt[chr(x)] = 0

# 알파벳 리스트 만들기
apb_list = []
for _ in range(n):
    apb_list.append(input())

# 알파벳 별 도합 차수 계산하기
for apb in apb_list:
    for i in range(len(apb)):
        apb_cnt[apb[i]] += 10**(len(apb)-i-1)
        
heap = []
for k,v in apb_cnt.items():
    if v != 0:
        heap.append(-v)

heapq.heapify(heap)
ans = 0
num = 9
while heap:
    tmp = heapq.heappop(heap)
    ans += num*tmp*(-1)
    num -= 1
    
print(ans)
```
