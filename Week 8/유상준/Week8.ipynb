# 문제 1
- 미로만들기
- https://www.acmicpc.net/problem/2665

```python
import sys
import heapq

n = int(input())
board = [list(map(int,input())) for _ in range(n)]

dx = [-1,0,1,0]
dy = [0,1,0,-1]
costs = [[sys.maxsize]*n for _ in range(n)]
q = []
heapq.heappush(q, (0,0,0)) # cost, x, y
costs[0][0] = 0
while q:
    cost, x, y = heapq.heappop(q)
    if cost > costs[x][y]:
        continue
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if 0<=nx<n and 0<=ny<n:
            n_cost = cost       # 이 부분 없이 그냥 cost 사용시 값이 달라짐 (for문 안에서 계속 +1 될 수도 있기 때문에)
            if board[nx][ny] == 1:
                pass
            else:
                n_cost += 1
            if n_cost < costs[nx][ny]:
                costs[nx][ny] = n_cost
                heapq.heappush(q, (n_cost,nx,ny))
            
print(costs[n-1][n-1])
```
